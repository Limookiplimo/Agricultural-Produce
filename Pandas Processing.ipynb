```python
# Pattern matching with regex
grains_regex = '|'.join(grains)
sea_food_regex = '|'.join(sea_food)
vegetables_regex = '|'.join(vegetables)
sugars_regex= '|'.join(sugars)
fruits_regex = '|'.join(fruits)
legumes_regex = '|'.join(legumes)
seeds_regex = '|'.join(seeds)
nuts_regex = '|'.join(nuts)
oils_fats_regex = '|'.join(oils_fats)
beverages_regex = '|'.join(beverages)
spices_regex = '|'.join(spices)
meat_regex = '|'.join(meat)
dairy_regex = '|'.join(dairy)
poultry_regex = '|'.join(poultry)
other_animal_products_regex = '|'.join(other_animal_products)
miscellaneous_regex = '|'.join(miscellaneous)
other_plant_products_regex = '|'.join(other_plant_products)
```

```python
df["Category"] = ""
df["Category"][df['Item'].str.contains(beverages_regex)] = "Beverages"
df["Category"][df['Item'].str.contains(grains_regex)] = "Grains"
df["Category"][df['Item'].str.contains(fruits_regex)] = "Fruits"
df["Category"][df['Item'].str.contains(vegetables_regex)] = "Vegetables"
df["Category"][df['Item'].str.contains(sugars_regex)] = "Sugars & Sweeteners"
df["Category"][df['Item'].str.contains(legumes_regex)] = "Legumes"
df["Category"][df['Item'].str.contains(seeds_regex)] = "Seeds"
df["Category"][df['Item'].str.contains(nuts_regex)] = "Nuts"
df["Category"][df['Item'].str.contains(oils_fats_regex)] = "Oils & Fats"
df["Category"][df['Item'].str.contains(spices_regex)] = "Spices"
df["Category"][df['Item'].str.contains(meat_regex)] = "Meat"
df["Category"][df['Item'].str.contains(sea_food_regex)] = "Seafood"
df["Category"][df['Item'].str.contains(dairy_regex)] = "Dairy"
df["Category"][df['Item'].str.contains(poultry_regex)] = "Poultry"
df["Category"][df['Item'].str.contains(other_animal_products_regex)] = "Other animal products"
df["Category"][df['Item'].str.contains(miscellaneous_regex)] = "Miscelaneous"
df["Category"][df['Item'].str.contains(other_plant_products_regex)] = "Other animal products"
```
